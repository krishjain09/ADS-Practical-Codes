```python
import pandas as pd
```

```python
df = pd.read_csv('loan_data_set.csv')
df.head(10)
```

```python
df.info()
```

```python
na_variables = []
for col in df.columns:
    missing_percentage = df[col].isna().mean()
    if missing_percentage > 0 :
        na_variables.append(col)
print(na_variables)
```

```python
df[['Gender','Married','Self_Employed']].info()
df[df[['Gender','Married','Self_Employed']].isnull().any(axis=1)]
```

```python
#mode imputation -> Replace missing values with most frequent value.
categorical_cols = ['Gender', 'Married','Self_Employed']
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    print(f"For {col} :  {mode_val}")
    for c in range(len(df[col])):
        if pd.isnull(df.loc[c,col]):
            df.loc[c,col] = mode_val
df[['Gender', 'Married','Self_Employed']].info()
```

```python
df.loc[11,"Self_Employed"]
df.head(25)
```

```python
df['Loan_Amount_Term'].describe()
```

```python
for i in range(len(df)):
    mean_value = df["Loan_Amount_Term"].mean()
    if pd.isnull(df.loc[i,"Loan_Amount_Term"]):
        df.loc[i,"Loan_Amount_Term"] = mean_value

df["Loan_Amount_Term"].info()
        
```

```python
for i in range(len(df)):
    mode_val = df['Credit_History'].mode()[0]
    if pd.isnull(df.loc[i,'Credit_History']):
        df.loc[i,'Credit_History'] = mode_val
df['Credit_History'].info()
```

```python
#Regression Imputation
from sklearn.linear_model import LinearRegression

train = df[df['LoanAmount'].notnull()]
# train.info()

test = df[df['LoanAmount'].isna()]
# test.info()

X_train = train[['CoapplicantIncome']]
Y_train = train['LoanAmount']

model = LinearRegression()
model.fit(X_train,Y_train)

print(df["LoanAmount"].info())

prediction = model.predict(test[['CoapplicantIncome']])
prediction = list(prediction)
print(prediction)
index = 0
for i in range(len(df)):
    if pd.isnull(df.loc[i,"LoanAmount"]):
        df.loc[i,"LoanAmount"]= prediction[index]
        index+=1

print(df.info())
```

```python
#Categorical → Numerical
from sklearn.preprocessin
g# import OrdinalEncoder
# TWO things happen here:

# 1. fit()
# 👉 Learn all unique categories
# 2. transform()
# 👉 Convert them into numbers

model = OrdinalEncoder()
categorical_cols = ['Gender', 'Married','Self_Employed','Property_Area','Loan_Status','Education']
for col in categorical_cols:
    df[col] = model.fit_transform(df[[col]])

df.head()
```

```python
#Random sample
df['Dependents'].info()
```

```python
nonnull_values = df['Dependents'].dropna()
print(nonnull_values)
nullcount = df['Dependents'].isna().sum()
print(nullcount)
```

```python
random = nonnull_values.sample(nullcount,random_state=0)
# print(random)
random = list(random)
print(random)

index = 0 ;
for i in range(len(df)):
    if pd.isnull(df.loc[i,"Dependents"]):
        df.loc[i,"Dependents"]= random[index]
        index+=1
print(df['Dependents'].info())
```
