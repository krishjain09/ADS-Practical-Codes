```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("supermarket_sales - Sheet1.csv")
print(df.head())
print(df.describe())
```

```python
plt.scatter(df['Tax 5%'],df['Unit price'],c='blue')
plt.show()
```

```python
plt.boxplot(df[['Unit price','Quantity' , 'Tax 5%']], tick_labels=['Unit price','Quantity' , 'Tax 5%'])
plt.show()
```

```python
plt.scatter(df['Total'], df['Tax 5%'], s=df['Quantity'])

plt.xlabel('Total Sales')        # X-axis name
plt.ylabel('Tax 5%')             # Y-axis name
plt.title('Bubble Chart: Total vs Tax (Size = Quantity)')

plt.show()
```

```python
df['Total'].plot.density()
plt.show()
```

```python
df['Total'].hist()
plt.show()
```

```python
import seaborn as sns
```

```python
#joint plot => shows relationship between 2 numerical variables along with their distribution
sns.jointplot(x=df['Quantity'], y=df['Total'], data=df)
```

```python
#pairplot => shows pairwise relationship between multiple variables
sns.pairplot(df.select_dtypes(include='number'))
```

```python
#bar chart => shows comparison between categories
df_copy = df['Gender'].value_counts()
print(df_copy)
df['Gender'].value_counts().plot.bar()
plt.show()
```

```python
#pie chart => shows proportion of categories as percentages
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
#Andrews Curves convert each data point into a curve using mathematical functions.
pd.plotting.andrews_curves(df[['Unit price','Quantity','Tax 5%','Gender']],'Gender')
plt.show()
```
